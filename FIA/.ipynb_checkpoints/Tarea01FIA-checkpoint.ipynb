```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as mat
```

```python
basedenombres = pd.read_csv('nombres.csv')
```

```python
basedenombres.info()
```

```python
basedenombres.head()
```

```python
basedenombres.head(50)
```

```python
nombremadeleineF = basedenombres[(basedenombres['sex']== 'F') & (basedenombres['name'] =='Madeleine')]
```

```python
media = nombremadeleineF['quantity'].mean()
moda = nombremadeleineF['quantity'].mode()
desviacion = nombremadeleineF['quantity'].std()
print("Media:", media)
print("Moda:", moda.values)
print("Desviación estándar:", desviacion)
```

```python
nombremadeleineM = basedenombres[(basedenombres['sex'] == 'M') & (basedenombres['name']
== 'Madeleine')]
```

```python
media = nombremadeleineM['quantity'].mean()
moda = nombremadeleineM['quantity'].mode()
desviacion = nombremadeleineM['quantity'].std()
print("Media:", media)
print("Moda:", moda.values)
print("Desviación estándar:", desviacion)
```

```python
total_mujeres = nombremadeleineF['quantity'].sum()
total_hombres = nombremadeleineM['quantity'].sum()
print("Total mujeres con el nombre de Madeleine:", total_mujeres)
print("Total hombres con el nombre de Madeleine:", total_hombres)
```

```python
madeleineF_x_estado = nombremadeleineF.groupby('state').agg({'quantity': 'sum'})
```

```python
madeleineF_x_estado
```

```python
madeleineM_x_estado = nombremadeleineM.groupby('state').agg({'quantity': 'sum'})
```

```python
madeleineM_x_estado
```

```python
mat.figure(figsize=(10,5))
mat.bar(madeleineF_x_estado.index, madeleineF_x_estado['quantity'], label='Mujeres', alpha=0.7)
mat.bar(madeleineM_x_estado.index, madeleineM_x_estado['quantity'], label='Hombres', alpha=0.7)
mat.xlabel('Estado')
mat.ylabel('Cantidad')
mat.title('Total de personas con el nombre Madeleine por estado')
mat.legend()
mat.xticks(rotation=90)
mat.show()
```
